```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
ca = pd.read_csv('data/ca.csv')
```

```python
# drop redundent rows
ca.drop(ca[ca['Track Name'] == 'Track Name'].index, inplace = True)
```

```python
# change columns data type
ca['Position'] = ca['Position'].astype('int32')
```

```python
rank_data = ca.groupby(['Track Name', 'Artist'], sort = False)\
               .agg({'region': 'count',
                    'Streams': 'max',
                    'Position': 'min',
                    'spotify_id': 'min'})\
                .reset_index()\
                .sort_values(['region'], ascending = False)\
                .reset_index()
```
